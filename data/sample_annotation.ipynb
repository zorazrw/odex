```python
import evaluate
code_eval_metric = evaluate.load("code_eval")

import os 
os.environ["HF_ALLOW_CODE_EVAL"] = "1"    # to enable execution 
```

### Sample \#1: Input Arguments Only

```python
# given in the google sheet 
rewritten_intent = "check if all elements in list `myList` are identical"
reference_snippet = "all(x == myList[0] for x in myList)"
```

```python
# to do: write the 'canonical_solution', 'prompt', 'suffix', and 'entry_point'
canonical_solution = "all(x == myList[0] for x in myList)"

# wrap the snippet into a function
input_args = ['myList']
entry_point = "wrapper_function"   # name of the wrapper function
prompt = f"def {entry_point}({', '.join(input_args)}):\n    return "

suffix = ""
wrapped_function = f"{prompt}{canonical_solution}{suffix}"
print(wrapped_function)
```

```python
# to do: write unit tests
test = """
def check(candidate):
    assert candidate([1,2,3]) == False
    assert candidate([1,1,1,1,1,1]) == True
    assert candidate([None,'%$#ga',3]) == False
"""
refernce_function = f"\n{test}\ncheck({entry_point})"
```

```python
# to do: verify the correctness of your annotated sample
code_eval_metric.compute(
    predictions=[[wrapped_function]], 
    references=[refernce_function], 
    k=[1,2], 
    num_workers=1, 
)   
# make sure your solution passes, and put it into the 'pass_check' field in the google sheet
```

### Sample \#2: With Input and Output Arguments

```python
# given in the google sheet 
rewritten_intent = "convert a string `my_string` with dot and comma into a float number `my_float`"
reference_snippet = "my_float = float(my_string.replace(',', ''))"
```

```python
# to do: write the canonical solution, wrap it into test function
canonical_solution = "my_float = float(my_string.replace(',', ''))"

# wrap the snippet into a function
input_args = ['my_string']
output_args = ['my_float']
entry_point = "wrapper_function"   # name of the wrapper function
prompt = f"def {entry_point}({', '.join(input_args)}):\n    "

suffix = f"\n    return {', '.join(output_args)}"
wrapped_function = f"{prompt}{canonical_solution}{suffix}"
print(wrapped_function)
```

```python
# to do: write unit tests
test = """
def check(candidate):
    assert (candidate('1,234.00') - 1234.0) < 1e-6
    assert (candidate('0.00') - 0.00) < 1e-6
    assert (candidate('1,000,000.00') - 999999.98) > 1e-6
    assert (candidate('1,000,000.00') - 1000000.00) < 1e-6
"""
refernce_function = f"\n{test}\ncheck({entry_point})"
```

```python
# to do: verify the correctness of your annotated sample
code_eval_metric.compute(
    predictions=[[wrapped_function]], 
    references=[refernce_function], 
    k=[1,2], 
    num_workers=1, 
)   
# make sure your solution passes, and put it into the 'pass_check' field in the google sheet
```
